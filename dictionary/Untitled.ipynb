### Implementing Own HashFunction
**Hashfuntion is Madeup of two funtion **
- **HashCode** - To conver out key into integer


- **Compression** Function > Our Hashcode will any integer but to make sure that it will come into the limit of my Bucket Size we use Compression Function & the                        perfect compression funtion shoud be %(modulo) of bucket size cz % b_s                         will always give the integer in the range from 0 to bucket_size-1.



#### Collision Handling
- If there are multiple keys passing from the hashfuntion have the same index in a memory(arr) So how can we store multiple keys which are reffering to the same index is what called as **Collision  Handling**.


#### To Handle Collision I am using Close Hashing
- It says store at the same index. Store multiple key value pair at the same index in the form of linked list. This is also known as **Seperate Chaining**.

**Impleted LoadFactor To maintain O(1) TC for the HashMethods.**
                        

```python
class MapNode:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None
class Map:
    def __init__(self):
        self.bucketSize = 5
        self.buckets = [ None for i in range(self.bucketSize)]
        self.count = 0
        
    def getBucketIndex(self,hc):
        return abs(hc%self.bucketSize)
    
    def size(self):
        return self.count
    
    def getValue(self,key):
        hc = hash(key)
        index = self.getBucketIndex(hc)
        head = self.buckets[index]
        while head is not None:
            if head.key == key:
                 return head.value
            head = head.next
        return None
    
    def remove(self,key):
        hc = hash(key)
        index = self.getBucketIndex(hc)
        head = self.buckets[index]
        prev = None
        while head is not None:
            if head.key == key:
                if prev is None:
                    self.buckets[index] = head.next
                else:
                    prev.next = head.next
                self.count -= 1
                return head.value
            prev = head
            head = head.next
        return None
    
    def rehash(self):
        temp = self.buckets
        self.buckets = [None for i in range(2*self.bucketSize)]
        self.bucketSize = 2*self.bucketSize
        self.count = 0
        for head in temp:
            while head is not None:
                self.insert(head.key,head.value)
                head = head.next
    
    def loadFactor(self):
        return self.count/self.bucketSize
    
    def insert(self, key, value):
        hc = hash(key)
        index = self.getBucketIndex(hc)
        head = self.buckets[index]
        while head is not None:
            if head.key == key: # if node is already present at the index just update the value and return
                head.value = value
                return
            head = head.next
            
        head = self.buckets[index]
        newNode = MapNode(key,value)
        newNode.next = head
        self.buckets[index] = newNode
        self.count += 1
        loadfactor = self.loadFactor()
        if loadfactor >= 0.7:
            print("as",loadfactor,">=0.7 we are going to change our bucket size from",self.bucketSize, "to 2x",self.bucketSize)
            self.rehash()
```

```python
def subsetSum(l):
    map = {}
    ans = 0
    sm = 0
    for idx,elem in enumerate(l):
        sm+=elem
        if map.get(sm) != None:
            ans = max(ans,idx-map[sm])
        else:
            map[sm] = idx

    return ans
            

# Main
n=int(input())
l=list(int(i) for i in input().strip().split(' '))
print(subsetSum(l))
```
